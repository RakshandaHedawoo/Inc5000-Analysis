# Inc 5000 Analysis
Inc 5000 (https://www.inc.com/inc5000) is a respected organization that annually compiles lists of high-growth companies.  
The data file contains information about companies that were included in the Inc. 5000 list for the year 2018. These companies were ranked based on their percentage revenue growth from 2014 to 2017. To be eligible for inclusion, these companies needed to meet specific criteria:  

* They must have been established and generating revenue by March 31, 2014.
* They must be privately held, for-profit entities based in the United States.
* They should operate independently, not as subsidiaries or divisions of other companies, as of December 31, 2017. (Note that some of the companies on the list may have since gone public or been acquired.)
* The minimum revenue required for 2014 was 100,000 dollars, and the minimum for 2017 was 2 million dollars.
* Growth rates used to rank these companies were calculated with precision to two decimal places. In cases where multiple companies had the same growth rate, those with higher revenue were given a higher ranking.  


#### DataSet Structure: inc 5000 company data
1. **Url (string)**  : Url link of inc 5000 website for each company located on inc 5000
1. **Rank (int)**  : Rank of the company according to the growth rate
1. **City (string)**  : City in which the company is located
1. **Growth (decimal)**  : Growth rate of the company
1. **Employees (int)**  : Number of employees in the company
1. **Company Name (string)**  : Company's name
1. **Website (string)**  : link for company's website
1. **City.1 (string)**  : State in which the company is located
1. **State (string)** : State Code of the state in which the company is located
1. **Revenuw (decimal)**  : Revenue generated by the company
1. **Zip Code (string)**  : Zip code of the place where the company is located
1. **Founded (int)**  : Founding year of the company
1. **Industry (string)**  : Different industry to which the company belongs
1. **City,State (string)**  : City and State in which the company is located

# Data Cleaning  
Before we start analysing the above data, let us see if the data needs some cleaning to do.  
Oh! there are some irregularities in the data that we must handle that are:
## Handling duplicates:
* In the data frame we can see that in the index, two companies have the same rank even though their growth rate is significantly different.
* Therefore, to handle this duplicate rank we will correct the ranks below 8. As, there was just one duplicate data it will be possible for me to correct it, however if their where multiple duplicates I would have to use some other column as the index for the data. 

```python
#finding the duplicate
inc_df.reset_index(inplace = True)
inc_df.duplicated(subset='RANK').head(10)
```

```python
#Fixing the duplicate by correcting the rank
inc_df['RANK'] = np.arange(1, len(inc_df) + 1)
inc_df.set_index('RANK',inplace=True)
inc_df.head()
```

## Data Transformation- Renaming and Deleting some columns: 
* Next, we can clearly see that some columns are misspelled for example 'REVENEW', 
* Secondly, we have a duplicate column name as CITY.1 which in fact is not a city name but a state name.
* Thirdly, we have some unnamed columns that have values only for the first 2 rows and rest of them are missing.
* So in order to proceed with the analysis we will first, rename the column names correctly, then change the 'CITY.1' column into 'STATE' and 'STATE' into 'STATE CODE', and delete the last two columns as they do not seem very useful for doing further analysis.  

Hence, the the final dataset should look like this:

```python
#renaming columns for better understanding
inc_df.rename(columns={'CITY.1':'State','STATE':'State Code','REVENUW':'Revenue','EMPLOYEES':'Employee Count','CITY':'City'
                      ,'GROWTH':'Growth','COMPANY NAME':'Company Name','WEBSITE':'Website','ZIP CODE':'Zip Code',
                       'FOUNDED':'Founding Year','INDUSTRY':'Industry','URL':'Url',},inplace = True)
#removing unwanted columns
inc_df.drop(['CITY, STATE','Unnamed: 14','Unnamed: 15'],axis=1,inplace=True)
inc_df.head()
```

# Analysing INC 5000 companies
Now that we have an overall idea about how the revenue data is distributed for different companies, let us analyse further according to different columns in the data.  
Instead of filtering the data again and again according to the different conditions, we can create a function that does the filtering work for us. The below function, returns us the result on basis of our query for a column and its value.

### Let us see some examples of using a custom-made filtering function:

```python
# we should list at least 20 rows
pd.set_option('min_rows',20)

# get's data based on a particular column name and a value for analysing column-wise data
def display_data(dataframe,number_of_rows=10,Category='None',value='None'):
    query_result = dataframe.sort_values(by='Revenue',ascending=False)
    if(Category!='None' and value!='None'):
        query_result = query_result[query_result[Category]==value]
    else:
        return query_result.iloc[:number_of_rows,:]
    return query_result.iloc[:number_of_rows,:]
```

# Awards Distribution
Now that we have analysed some data column-wise, INC 5000 wants to present two awards to the Service-based companies listed in the top 50.  
1. **Trailblazer Award** : Given to the companies that are under top 25
1. **Pioneer Award** : Given to the companies that didn't qualify for top 25, however are under top 50.

Firstly, we will add another column in our dataset called as 'Awards'.  
Then we will assign the value of 'Trailblazer' to the top 25 companies and 'Pioneer' to the companies not in top 25 but in top 50.  

### Let us have a look at which companies got Trailblazers and which of them received Pioneer award

```python
#Filtering companies by Service industry
service_based_company = inc_df[inc_df['Industry'].str.contains('Services')].sort_values(by='Revenue',ascending=False)
service_based_company.reset_index(inplace=True)

# additional Award column
service_based_company['Awards']='No Award'

# top quarter company count
top_quarter_count = int(service_based_company['Company Name'].count()*25/100)
# top half company count
top_half_count = int(service_based_company['Company Name'].count()*50/100)

# Trailblazer award to top quarter companies
service_based_company.loc[:top_quarter_count-1,'Awards'] = 'Trailblazer'

# Pioneer award to companies in top 50 but not in top 25
service_based_company.loc[top_quarter_count:top_half_count-1,'Awards'] = 'Pioneer'

service_based_company.loc[service_based_company['Awards']!='No Award',('Company Name','RANK','Revenue','Awards')]
```
